````python
#Tugas
#Aplikasi jadwal piket satpam
#Sebuah perusahaan memiliki 4 satpam yang bertugas menjaga keamanan perusahaan setiap
#hari. Jadwal piket mereka harus bergantian setiap hari dengan aturan rotasi tetap.
#Anda diminta untuk membuat program yang mencetak jadwal piket satpam selama 2 minggu (14 hari).
#Ketentuan Program:
#1. Pengguna dapat memasukkan 4 nama satpam yang akan bertugas.
#2. Nama satpam harus berputar secara otomatis menggunakan Circular Linked List.
#3. Setiap satpam yang bertugas akan kembali bertugas setelah 4 hari.
#4. Program harus mencetak jadwal lengkap selama 2 minggu (14 hari)

#Ketentuan pelaporan tugas:
#1. Saat menjalankan program, masukkan nama Anda, dan 3 nama teman dengan NPM sebelum atau sesudah Anda!
#2. Simpan source code project Anda dan unggah juga pada GitHub!
#3. Masukkan Tugas ini ke dalam laporan praktikum pada Medium dan berikan ulasan yang lengkap mengenai jalannya program ini serta penjelasan mengenai implementasi
#Circular Linked List secara lengkap pada program ini!
#4. Jangan lupa sertakan screenshot dari hasil jalannya program.


class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nama satpam
        self.next = None  # Pointer ke node berikutnya

class CircularLinkedList:
    def __init__(self):
        self.head = None  # Awalnya kosong

    def add_satpam(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head  # Circular
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def print_schedule(self, days=14):
        if not self.head:
            print("Tidak ada satpam yang bertugas.")
            return

        print("\nJadwal Piket Satpam Selama 2 Minggu:\n")
        temp = self.head
        for day in range(1, days + 1):
            print(f"Hari {day}: {temp.data}")
            temp = temp.next

# Program Utama
satpam_list = CircularLinkedList()
print("Masukkan nama 4 satpam yang akan bertugas:")
for i in range(4):
    nama = input(f"Masukkan nama satpam ke-{i+1}: ")
    satpam_list.add_satpam(nama)

satpam_list.print_schedule()
````

Circular Linked List adalah struktur data yang mirip dengan Singly Linked List, tetapi node terakhir akan terhubung kembali ke node pertama, membentuk lingkaran. Dalam program ini, Circular Linked List digunakan untuk memastikan bahwa rotasi piket tetap berjalan secara berulang dan otomatis.

1. Struktur Node dan Circular Linked List
  ```
  class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nama satpam
        self.next = None  # Pointer ke node berikutnya
  ```
  Setiap Node menyimpan:
  * data → Nama satpam
  * next → Pointer ke satpam berikutnya dalam daftar
  Kemudian, Circular Linked List didefinisikan dengan head sebagai awal daftar:
  ```
  class CircularLinkedList:
    def __init__(self):
        self.head = None  # Awalnya kosong
  ```

2. Menambahkan Satpam ke Circular Linked List
  Metode add_satpam(data) digunakan untuk menambahkan satpam ke dalam daftar:
  ```
  def add_satpam(self, data):
    new_node = Node(data)
    if not self.head:
        self.head = new_node
        self.head.next = self.head  # Circular
    else:

        temp.next = new_node

  ```
  Penjelasan:
  * Jika Circular Linked List masih kosong, maka:
    * head akan menunjuk ke node pertama.
    * Node pertama akan menunjuk ke dirinya sendiri (new_node.next = self.head).
  * Jika sudah ada satpam dalam daftar, maka:
    * Node baru ditambahkan di akhir daftar.
    * Node terakhir akan menunjuk kembali ke head, sehingga tetap circular.

  Penjelasan:

    * Pada setiap hari, mencetak nama satpam yang bertugas.
    * Pointer temp akan berpindah ke node berikutnya menggunakan temp = temp.next, yang memastikan rotasi berjalan otomatis.
