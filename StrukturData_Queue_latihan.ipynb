```python
#Latihan / Tugas 1

class QueueNode:
    def __init__(self, ticket_id):
        self.ticket_id = ticket_id
        self.next = None

class BioskopQueue:
    def __init__(self, max_capacity=10):
        self.max_capacity = max_capacity
        self.head = None
        self.tail = None
        self.count = 0
        self.customer_code = "AN21"  # dua huruf pertama nama (AN) + dua digit terakhir NPM (21)
        self.ticket_counter = 0
        self.waiting_list = []

    def generate_ticket_id(self):
        self.ticket_counter += 1
        return f"{self.customer_code}-{self.ticket_counter}"

    def is_empty(self):
        return self.head is None

    def size(self):
        return self.count

    def is_vip(self, ticket_id):
        return '7' in ticket_id

    def enqueue(self, ticket_id=None):
        if ticket_id is None:
            ticket_id = self.generate_ticket_id()

        # Check if customer is VIP
        if self.is_vip(ticket_id):
            print(f"VIP Customer {ticket_id} mendapatkan kursi tanpa antri!")
            return ticket_id

        # If queue is full, add to waiting list
        if self.count >= self.max_capacity:
            self.waiting_list.append(ticket_id)
            print(f"Antrian penuh! {ticket_id} masuk ke daftar tunggu.")
            return ticket_id

        new_node = QueueNode(ticket_id)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

        self.count += 1
        print(f"Pelanggan {ticket_id} masuk ke antrian. Posisi: {self.count}")
        return ticket_id

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        ticket_id = self.head.ticket_id

        self.head = self.head.next

        if self.head is None:
            self.tail = None

        self.count -= 1
        print(f"Pelanggan {ticket_id} sedang dilayani.")

        if self.waiting_list and self.count < self.max_capacity:
            waiting_customer = self.waiting_list.pop(0)
            self.enqueue(waiting_customer)
            print(f"Pelanggan {waiting_customer} dipindahkan dari daftar tunggu ke antrian.")

        return ticket_id

    def front(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None
        return self.head.ticket_id

    def rear(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None
        return self.tail.ticket_id

    def display_queue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return

        current = self.head
        print("Antrian saat ini:")
        position = 1

        while current:
            print(f"{position}. {current.ticket_id}")
            current = current.next
            position += 1

        if self.waiting_list:
            print("\nDaftar tunggu:")
            for i, customer in enumerate(self.waiting_list, 1):
                print(f"{i}. {customer}")

def main():
    bioskop = BioskopQueue(max_capacity=5)

    print("=== SISTEM ANTRIAN PEMESANAN TIKET BIOSKOP ===")

    print("\n1. Menambahkan pelanggan ke antrian:")
    bioskop.enqueue()
    bioskop.enqueue()
    bioskop.enqueue()

    print("\n2. Menampilkan antrian:")
    bioskop.display_queue()

    print("\n3. Pelanggan pertama dan terakhir:")
    print(f"Front: {bioskop.front()}")
    print(f"Rear: {bioskop.rear()}")

    print("\n4. Melayani pelanggan pertama:")
    bioskop.dequeue()
    bioskop.display_queue()

    print("\n5. Informasi antrian:")
    print(f"Jumlah pelanggan dalam antrian: {bioskop.size()}")
    print(f"Antrian kosong: {bioskop.is_empty()}")

    print("\n6. Test kapasitas maksimal:")
    bioskop.enqueue()
    bioskop.enqueue()
    bioskop.enqueue()
    bioskop.enqueue()
    bioskop.display_queue()

    print("\n7. Test VIP customer:")
    vip_id = bioskop.customer_code + "-7"
    bioskop.enqueue(vip_id)

    print("\n8. Melayani beberapa pelanggan:")
    bioskop.dequeue()
    bioskop.dequeue()
    bioskop.display_queue()

if __name__ == "__main__":
    main()
```

```python
#Latihan / Tugas 2

class QueueNode:
    def __init__(self, order_id, num_pages):
        self.order_id = order_id
        self.num_pages = num_pages
        self.next = None
        self.is_second_phase = False
class FotokopiQueue:
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0
        self.customer_prefix = "ANA21"  # tiga huruf pertama nama (ANA) + dua digit terakhir NPM (21)
        self.order_counter = 0

    def generate_order_id(self):
        self.order_counter += 1
        return f"{self.customer_prefix}-{self.order_counter}"

    def is_empty(self):
        return self.head is None

    def size(self):
        return self.count

    def is_fast_track(self, num_pages):
        return num_pages < 10

    def enqueue(self, order_id=None, num_pages=0, is_second_phase=False):
        if order_id is None:
            order_id = self.generate_order_id()

        if self.is_fast_track(num_pages) and not is_second_phase:
            print(f"Fast Track! Pesanan {order_id} ({num_pages} halaman) langsung diproses tanpa mengantri.")
            return order_id

        new_node = QueueNode(order_id, num_pages)
        new_node.is_second_phase = is_second_phase

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

        self.count += 1

        if is_second_phase:
            print(f"Pesanan {order_id} (tahap kedua, {num_pages} halaman) masuk ke antrian. Posisi: {self.count}")
        else:
            print(f"Pesanan {order_id} ({num_pages} halaman) masuk ke antrian. Posisi: {self.count}")

        return order_id

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        order_id = self.head.order_id
        num_pages = self.head.num_pages
        is_second_phase = self.head.is_second_phase

        self.head = self.head.next

        if self.head is None:
            self.tail = None

        self.count -= 1

        if num_pages > 100 and not is_second_phase:
            first_phase_pages = num_pages // 2
            second_phase_pages = num_pages - first_phase_pages

            print(f"Pesanan besar {order_id} ({num_pages} halaman) diproses tahap pertama ({first_phase_pages} halaman).")
            self.enqueue(order_id, second_phase_pages, True)
        else:
            if is_second_phase:
                print(f"Pesanan {order_id} tahap kedua ({num_pages} halaman) selesai diproses.")
            else:
                print(f"Pesanan {order_id} ({num_pages} halaman) selesai diproses.")

        return order_id

    def front(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        status = "tahap kedua" if self.head.is_second_phase else ""
        return f"{self.head.order_id} ({self.head.num_pages} halaman) {status}".strip()

    def rear(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        status = "tahap kedua" if self.tail.is_second_phase else ""
        return f"{self.tail.order_id} ({self.tail.num_pages} halaman) {status}".strip()

    def display_queue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return

        current = self.head
        print("Antrian fotokopi saat ini:")
        position = 1

        while current:
            status = "tahap kedua" if current.is_second_phase else ""
            print(f"{position}. {current.order_id} ({current.num_pages} halaman) {status}".strip())
            current = current.next
            position += 1

def main():
    fotokopi = FotokopiQueue()

    print("=== SISTEM ANTRIAN LAYANAN FOTOKOPI KAMPUS ===")

    print("\n1. Menambahkan beberapa pesanan fotokopi:")
    fotokopi.enqueue(num_pages=50)
    fotokopi.enqueue(num_pages=5)
    fotokopi.enqueue(num_pages=120)
    fotokopi.enqueue(num_pages=30)

    print("\n2. Menampilkan antrian:")
    fotokopi.display_queue()

    print("\n3. Pesanan pertama dan terakhir:")
    print(f"Front: {fotokopi.front()}")
    print(f"Rear: {fotokopi.rear()}")

    print("\n4. Memproses beberapa pesanan:")
    fotokopi.dequeue()
    fotokopi.dequeue()
    fotokopi.display_queue()

    print("\n5. Informasi antrian:")
    print(f"Jumlah pesanan dalam antrian: {fotokopi.size()}")
    print(f"Antrian kosong: {fotokopi.is_empty()}")

    print("\n6. Memproses semua pesanan sampai antrian kosong:")
    while not fotokopi.is_empty():
        fotokopi.dequeue()

    print(f"Antrian kosong: {fotokopi.is_empty()}")

if __name__ == "__main__":
    main()
```

```python
#Latihan / Tugas 3

class QueueNode:
    def __init__(self, vehicle_number, is_electric=False):
        self.vehicle_number = vehicle_number
        self.is_electric = is_electric
        self.next = None

class ParkingQueue:
    def __init__(self, max_capacity=15):
        self.max_capacity = max_capacity
        self.head = None
        self.tail = None
        self.count = 0
        self.plate_prefix = "JA21"  # dua huruf pertama kota lahir (JA) + dua digit terakhir NPM (21)
        self.vehicle_counter = 0
        self.waiting_list = []
        self.priority_waiting = []

    def generate_vehicle_number(self):
        self.vehicle_counter += 1
        return f"{self.plate_prefix}-{self.vehicle_counter}"

    def is_empty(self):
        return self.head is None

    def is_full(self):
        return self.count >= self.max_capacity

    def size(self):
        return self.count

    def enqueue(self, vehicle_number=None, is_electric=False):
        if vehicle_number is None:
            vehicle_number = self.generate_vehicle_number()

        if self.is_full():
            if is_electric:
                self.priority_waiting.append((vehicle_number, is_electric))
                print(f"Parkiran penuh! Kendaraan listrik {vehicle_number} masuk ke daftar tunggu prioritas.")
            else:
                self.waiting_list.append((vehicle_number, is_electric))
                print(f"Parkiran penuh! Kendaraan {vehicle_number} masuk ke daftar tunggu.")
            return vehicle_number

        if is_electric and not self.is_empty():
            print(f"Kendaraan listrik {vehicle_number} mendapat parkir prioritas!")
            new_node = QueueNode(vehicle_number, is_electric)

            if self.count == 1:
                new_node.next = None
                self.head.next = new_node
                self.tail = new_node
            else:
                new_node.next = self.head.next
                self.head.next = new_node

            self.count += 1
            return vehicle_number

        new_node = QueueNode(vehicle_number, is_electric)

        if self.is_empty():
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

        self.count += 1

        if is_electric:
            print(f"Kendaraan listrik {vehicle_number} masuk ke parkiran. Jumlah kendaraan: {self.count}/{self.max_capacity}")
        else:
            print(f"Kendaraan {vehicle_number} masuk ke parkiran. Jumlah kendaraan: {self.count}/{self.max_capacity}")

        return vehicle_number

    def dequeue(self):
        if self.is_empty():
            print("Parkiran kosong!")
            return None

        vehicle_number = self.head.vehicle_number
        is_electric = self.head.is_electric

        self.head = self.head.next

        if self.head is None:
            self.tail = None

        self.count -= 1

        if is_electric:
            print(f"Kendaraan listrik {vehicle_number} keluar dari parkiran.")
        else:
            print(f"Kendaraan {vehicle_number} keluar dari parkiran.")

        # Check waiting list, priority first then regular
        if self.priority_waiting:
            waiting_vehicle, is_elec = self.priority_waiting.pop(0)
            self.enqueue(waiting_vehicle, is_elec)
            print(f"Kendaraan listrik {waiting_vehicle} dari daftar tunggu prioritas masuk ke parkiran.")
        elif self.waiting_list:
            waiting_vehicle, is_elec = self.waiting_list.pop(0)
            self.enqueue(waiting_vehicle, is_elec)
            print(f"Kendaraan {waiting_vehicle} dari daftar tunggu masuk ke parkiran.")

        return vehicle_number

    def front(self):
        if self.is_empty():
            print("Parkiran kosong!")
            return None
        return self.head.vehicle_number

    def rear(self):
        if self.is_empty():
            print("Parkiran kosong!")
            return None
        return self.tail.vehicle_number

    def display_parking(self):
        if self.is_empty():
            print("Parkiran kosong!")
            return

        current = self.head
        print(f"Kendaraan di parkiran ({self.count}/{self.max_capacity}):")
        position = 1

        while current:
            vehicle_type = "Listrik" if current.is_electric else "Regular"
            print(f"{position}. {current.vehicle_number} ({vehicle_type})")
            current = current.next
            position += 1

        if self.priority_waiting or self.waiting_list:
            print("\nDaftar tunggu:")

            if self.priority_waiting:
                print("Kendaraan prioritas:")
                for i, (vehicle, _) in enumerate(self.priority_waiting, 1):
                    print(f"{i}. {vehicle} (Listrik)")

            if self.waiting_list:
                print("Kendaraan regular:")
                for i, (vehicle, is_elec) in enumerate(self.waiting_list, 1):
                    vehicle_type = "Listrik" if is_elec else "Regular"
                    print(f"{i}. {vehicle} ({vehicle_type})")

def main():
    mall_parking = ParkingQueue(max_capacity=5)

    print("=== SISTEM ANTRIAN PARKIR MALL ===")

    print("\n1. Menambahkan beberapa kendaraan ke parkiran:")
    mall_parking.enqueue()
    mall_parking.enqueue()
    mall_parking.enqueue(is_electric=True)

    print("\n2. Menampilkan kendaraan di parkiran:")
    mall_parking.display_parking()

    print("\n3. Kendaraan pertama dan terakhir di parkiran:")
    print(f"Front: {mall_parking.front()}")
    print(f"Rear: {mall_parking.rear()}")

    print("\n4. Test kapasitas maksimal parkiran:")
    mall_parking.enqueue()
    mall_parking.enqueue()
    mall_parking.enqueue()
    mall_parking.enqueue(is_electric=True)
    mall_parking.display_parking()

    print("\n5. Mengeluarkan beberapa kendaraan:")
    mall_parking.dequeue()
    mall_parking.dequeue()
    mall_parking.display_parking()

    print("\n6. Informasi parkiran:")
    print(f"Jumlah kendaraan dalam parkiran: {mall_parking.size()}")
    print(f"Parkiran kosong: {mall_parking.is_empty()}")
    print(f"Parkiran penuh: {mall_parking.is_full()}")

    print("\n7. Test parkir prioritas (kendaraan listrik):")
    new_electric = mall_parking.enqueue(is_electric=True)
    mall_parking.display_parking()

    print("\n8. Mengeluarkan semua kendaraan:")
    while not mall_parking.is_empty():
        mall_parking.dequeue()

    print(f"Parkiran kosong: {mall_parking.is_empty()}")

if __name__ == "__main__":
    main()
```

Cara Kerja Code:

Kendaraan Biasa:
- Kendaraan masuk ke parkiran mengikuti urutan FIFO (First In, First Out).

Kendaraan Listrik:
- Diberikan prioritas dalam parkiran, ditempatkan setelah kendaraan yang masuk pertama.
- Memiliki daftar tunggu prioritas terpisah jika seluruh parkiran telah terisi.

Manajemen Parkir:
- Saat kendaraan keluar, kendaraan dari daftar tunggu akan segera masuk, dengan prioritas untuk kendaraan listrik terlebih dahulu.
- Status kapasitas parkiran akan ditampilkan setiap kali operasi dilakukan.
- Kendaraan biasa dan kendaraan listrik akan dipisahkan dan dilacak secara terpisah.

```python
#Latihan / Tugas 4

class QueueNode:
    def __init__(self, student_id, has_achievements=False):
        self.student_id = student_id
        self.has_achievements = has_achievements
        self.next = None

class CertificateQueue:
    def __init__(self, split_threshold=20):
        self.regular_head = None
        self.regular_tail = None
        self.achievement_head = None
        self.achievement_tail = None
        self.regular_count = 0
        self.achievement_count = 0
        self.split_threshold = split_threshold
        self.queue_prefix = "T21"  # huruf pertama program studi (T) + dua digit terakhir NPM (21)
        self.student_counter = 0
        self.is_split = False

    def generate_student_id(self):
        self.student_counter += 1
        return f"{self.queue_prefix}-{self.student_counter}"

    def is_empty(self):
        return self.regular_head is None and self.achievement_head is None

    def size(self):
        return self.regular_count + self.achievement_count

    def check_split_queue(self):
        if not self.is_split and self.size() > self.split_threshold:
            self.is_split = True
            print(f"Jumlah mahasiswa melebihi {self.split_threshold}! Antrian dibagi menjadi jalur reguler dan jalur mahasiswa berprestasi.")

    def enqueue(self, student_id=None, has_achievements=False, is_national_competitor=False):
        if student_id is None:
            student_id = self.generate_student_id()

        if is_national_competitor:
            print(f"Mahasiswa {student_id} (Peserta Kompetisi Nasional) dapat langsung mengambil sertifikat tanpa mengantri!")
            return student_id

        new_node = QueueNode(student_id, has_achievements)

        self.check_split_queue()

        if self.is_split and has_achievements:
            if self.achievement_head is None:
                self.achievement_head = new_node
                self.achievement_tail = new_node
            else:
                self.achievement_tail.next = new_node
                self.achievement_tail = new_node

            self.achievement_count += 1
            print(f"Mahasiswa {student_id} (Berprestasi) masuk ke jalur mahasiswa berprestasi. Posisi: {self.achievement_count}")
        else:
            if self.regular_head is None:
                self.regular_head = new_node
                self.regular_tail = new_node
            else:
                self.regular_tail.next = new_node
                self.regular_tail = new_node

            self.regular_count += 1

            if self.is_split:
                print(f"Mahasiswa {student_id} masuk ke jalur reguler. Posisi: {self.regular_count}")
            else:
                print(f"Mahasiswa {student_id} masuk ke antrian. Posisi: {self.regular_count}")

        return student_id

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        if self.is_split:
            if self.achievement_head is not None:
                student_id = self.achievement_head.student_id
                self.achievement_head = self.achievement_head.next

                if self.achievement_head is None:
                    self.achievement_tail = None

                self.achievement_count -= 1
                print(f"Mahasiswa {student_id} (Berprestasi) telah mengambil sertifikat.")
                return student_id
            else:
                student_id = self.regular_head.student_id
                self.regular_head = self.regular_head.next

                if self.regular_head is None:
                    self.regular_tail = None

                self.regular_count -= 1
                print(f"Mahasiswa {student_id} telah mengambil sertifikat.")
                return student_id
        else:
            student_id = self.regular_head.student_id
            self.regular_head = self.regular_head.next

            if self.regular_head is None:
                self.regular_tail = None

            self.regular_count -= 1
            print(f"Mahasiswa {student_id} telah mengambil sertifikat.")
            return student_id

    def front(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        if self.is_split:
            if self.achievement_head is not None:
                return f"{self.achievement_head.student_id} (Berprestasi)"
            else:
                return self.regular_head.student_id
        else:
            return self.regular_head.student_id

    def rear(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None

        if self.is_split:
            if self.achievement_tail is not None:
                return f"{self.achievement_tail.student_id} (Berprestasi)"
            else:
                return self.regular_tail.student_id
        else:
            return self.regular_tail.student_id

    def display_queue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return

        if self.is_split:
            if self.achievement_head is not None:
                current = self.achievement_head
                print(f"\nJalur Mahasiswa Berprestasi ({self.achievement_count} mahasiswa):")
                position = 1

                while current:
                    print(f"{position}. {current.student_id}")
                    current = current.next
                    position += 1

            if self.regular_head is not None:
                current = self.regular_head
                print(f"\nJalur Reguler ({self.regular_count} mahasiswa):")
                position = 1

                while current:
                    print(f"{position}. {current.student_id}")
                    current = current.next
                    position += 1
        else:
            current = self.regular_head
            print(f"\nAntrian Pengambilan Sertifikat ({self.regular_count} mahasiswa):")
            position = 1

            while current:
                print(f"{position}. {current.student_id}")
                current = current.next
                position += 1

def main():
    certificate_queue = CertificateQueue(split_threshold=5)

    print("=== SISTEM ANTRIAN PENGAMBILAN SERTIFIKAT KAMPUS ===")

    print("\n1. Menambahkan beberapa mahasiswa ke antrian:")
    certificate_queue.enqueue()
    certificate_queue.enqueue()
    certificate_queue.enqueue(has_achievements=True)
    certificate_queue.enqueue()

    print("\n2. Menampilkan antrian saat ini:")
    certificate_queue.display_queue()

    print("\n3. Mahasiswa pertama dan terakhir dalam antrian:")
    print(f"Front: {certificate_queue.front()}")
    print(f"Rear: {certificate_queue.rear()}")

    print("\n4. Test pembagian antrian (threshold):")
    certificate_queue.enqueue()
    certificate_queue.enqueue()
    certificate_queue.enqueue(has_achievements=True)
    certificate_queue.display_queue()

    print("\n5. Test mahasiswa kompetisi nasional:")
    certificate_queue.enqueue(is_national_competitor=True)

    print("\n6. Melayani beberapa mahasiswa:")
    certificate_queue.dequeue()
    certificate_queue.dequeue()
    certificate_queue.dequeue()
    certificate_queue.display_queue()

    print("\n7. Informasi antrian:")
    print(f"Jumlah total mahasiswa dalam antrian: {certificate_queue.size()}")
    print(f"Antrian kosong: {certificate_queue.is_empty()}")

    print("\n8. Melayani semua mahasiswa sampai antrian kosong:")
    while not certificate_queue.is_empty():
        certificate_queue.dequeue()

    print(f"Antrian kosong: {certificate_queue.is_empty()}")

if __name__ == "__main__":
    main()
```



# Struktur Antrian:
- Kami menggunakan implementasi linked list melalui kelas `QueueNode`.
- Terdapat antrian terpisah untuk mahasiswa reguler dan mahasiswa berprestasi jika diperlukan.

# Metode yang Digunakan:
- `enqueue(student_id)` – digunakan untuk menambahkan mahasiswa ke dalam antrian pengambilan sertifikat.
- `dequeue()` – berfungsi untuk melayani mahasiswa berikutnya dalam antrian.
- `front()` dan `rear()` – menampilkan mahasiswa yang pertama dan terakhir dalam antrian.
- `is_empty()` dan `size()` – memeriksa apakah antrian kosong dan menghitung jumlah mahasiswa yang ada.

# Fitur Khusus:
- ID mahasiswa mengikuti format "T21" (menggunakan huruf pertama dari program studi *Teknologi Informasi* ditambah dua digit terakhir NPM, yaitu "21").
- Antrian akan dipisahkan jika jumlah mahasiswa melebihi batas (default 20 mahasiswa, sedangkan dalam demo diatur menjadi 5 untuk keperluan pengujian).
- Mahasiswa berprestasi mendapatkan layanan prioritas ketika antrian terpisah.
- Mahasiswa yang mengikuti kompetisi tingkat nasional akan dilayani langsung tanpa harus antre.

# Operasi Awal:

- Ketika jumlah mahasiswa melebihi batas:
- Mahasiswa reguler akan masuk ke antrian reguler.
- Mahasiswa berprestasi akan dimasukkan ke dalam antrian prioritas.

# Penanganan Prioritas:
- Peserta kompetisi nasional akan dilayani secara langsung tanpa mengalami antrean.
- Jika antrian dipisahkan, mahasiswa dalam antrian berprestasi akan dilayani terlebih dahulu.
