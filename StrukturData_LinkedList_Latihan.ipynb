````python
#Implementasi Singly Linked List

class SinglyLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  def display(self):
    temp = self.head
    while temp:
      print(temp.data, end=" -> ")
      temp = temp.next
    print("None")
````

````python
#Implementasi Doubly Linked List

class DoublyNode:
  def __init__(self, data):
    self.data = data
    self.prev = None
    self.next = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = DoublyNode(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node
    new_node.prev = temp

  def display(self):
    temp = self.head
    while temp:
      print(temp.data, end=" -> ")
      temp = temp.next
    print("None")
````

````python
#Implementasi Circular Linked List

class CircularLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      new_node.next = self.head
      return
    temp = self.head
    while temp.next != self.head:
      temp = temp.next
    temp.next = new_node
    new_node.next = self.head

  def display(self):
    if not self.head:
      return
    temp = self.head
    while True:
      print(temp.data, end=" -> ")
      temp = temp.next
      if temp == self.head:
        break
    print("(Back to Head)")
````

````python
# Implementasi Doubly Linked Link

#Definisi node dalam doubly linked list
class PageNode:
  def __init__(self, url):
    self.url = url #URL halaman web
    self.next = None #Pointer ke halaman selanjutnya
    self.prev = None #Pointer ke halaman sebelumnya

# Definisi Doubly Linked  List untuk Riwayat Peramban
class BrowserHistory:
  def __init__(self):
    self.current = None #Pointer ke halaman yang sedang dibuka

  #Menambahkan halaman baru kedalam riwayat
  def visit_page(self, url): # Corrected indentation
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page #Jika Riwayat kosong, halaman inimenjadi halaman pertama
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page #Pindahkan posisi ke halaman baru

  #Menampilkan riwayat kunjungan
  def show_history(self): # Corrected indentation
    if not self.current:
      print ("\n Riwayat kosong.")
      return

    print("\n Riwayat Peramban Web:") # Corrected indentation
    temp = self.current
    while temp.prev: #Kembali ke halaman pertama
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.url}")
      temp = temp.next
      count += 1

  #Navigasi mundur ke halaman sebelumnya
  def go_back(self): # Corrected indentation
    if self.current and self.current.prev:
      self.current = self.current.prev # Corrected: sel.current to self.current
      print(f"\n Anda kembali ke: {self.current.url}")
    else:
      print("\n Tidak bisa kembali, ini adalah halaman pertama")

  #Navigasi maju kehalaman berikutnya
  def go_forward(self): # Corrected indentation
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.url}") # Added print statement to show the current URL
    else:
      print ("\n Tidak bisa maju, ini adalah halaman terakir")


# ========== Program Utama (Main) ========== #

# Membuat objek riwayat peramban
history = BrowserHistory()

# Meminta pengguna memasukan jumlah halaman yang dikunjungi
num_pages = int(input("Masukkan jumlah halaman yang dikunjungi: "))

# Loop untuk memasukan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)

# Menampilkan riwayat yang telah dimasukan
history.show_history()

#Menu navigasi pengguna
while True:
  command = input("\n Tekan'b' untuk kembali, tekan 'f' untuk maju, tean 'x' untuk keluar.\n Masukan perintah:").lower()
  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat prambanan.")
    break # Added break to exit the loop
  else:
    print("\n Perintah todak dikenal")
````

````python
#Latihan 2
# Definisi node dalam doubly linked list
class FolderNode:
    def __init__(self, name):
        self.name = name  # Nama folder
        self.next = None  # Pointer ke folder berikutnya
        self.prev = None  # Pointer ke folder sebelumnya

# Definisi Doubly Linked List untuk Riwayat Navigasi Folder
class FolderHistory:
    def __init__(self):
        self.current = None  # Pointer ke folder yang sedang dibuka

    # Menambahkan folder baru ke dalam riwayat
    def visit_folder(self, name):
        new_folder = FolderNode(name)
        if not self.current:
            self.current = new_folder  # Jika riwayat kosong, ini jadi folder pertama
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder  # Pindah ke folder baru

    # Menampilkan riwayat navigasi folder
    def show_history(self):
        if not self.current:
            print("\nRiwayat kosong.")
            return

        print("\nRiwayat Navigasi Folder:")
        temp = self.current
        while temp.prev:  # Kembali ke folder pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.name}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke folder sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke folder: {self.current.name}")
        else:
            print("\nTidak bisa kembali, ini adalah folder pertama.")

    # Navigasi maju ke folder berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke folder: {self.current.name}")
        else:
            print("\nTidak bisa maju, ini adalah folder terakhir.")

# ========== Program Utama (Main) ========== #

# Membuat objek riwayat navigasi folder
history = FolderHistory()

# Meminta pengguna memasukkan jumlah folder yang dikunjungi (sesuai digit terakhir NPM)
num_folders = 1  # Dari digit terakhir NPM 1010101021 (yakni 1)

# Memasukkan nama folder ke dalam Doubly Linked List
for i in range(1, num_folders + 1):
    folder_name = input(f"Masukkan nama folder ke-{i}: ")
    history.visit_folder(folder_name)

# Menampilkan riwayat folder yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
    command = input("\nTekan 'b' untuk kembali, tekan 'f' untuk maju, tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\nKeluar dari riwayat navigasi.")
        break
    else:
        print("\nPerintah tidak dikenal.")
````

````python
#Implementasi Circular Linked List
#Definnisi Node dalam circular Linked List
class SongNode:
  def __init__(self, titel):
    self.titel = titel #Menyimpan judul lagu
    self.next = None #Pointer ke lagu berikutnya

#Definisi circular linked list untuk playlist lagu
class Playlist:
  def __init__(self):
    self.current = None #Lagu yang sedang diputar

    #Menambahkan lagu kedalam playlist
  def add_song(self, titel):
    new_song = SongNode(titel)
    if not self.current:
      self.current = new_song
      self.current.next = self.current #Menghubungkan ke dirinya sendiri
    else:
      temp = self.current
      while temp.next != self.current: #Mencari lagu terakhir
          temp = temp.next
      temp.next = new_song #Menambahkan lagu diakhir
      new_song.next = self.current

  #Menampilkan daftar lagu dalam playlist
  def show_playlist(self):
    if not self.current:
      print("\n Playlist kosong.")
      return

    print("\n Playlist Lagu Anda:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.titel}")
      temp = temp.next
      count += 1
      if temp == self.current: #Berhenti jika sudah kembali ke awal
        break

  #Memutar lagu berikutnya dalam playlist
  def play_next(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar: {self.current.titel}")

# ========== Program Utama Main ========== #

#Membuat Objek playlist
playlist = Playlist()

#Meminta pengguna memasukkan jumlah lagu
num_songs = int(input("Masukkan jumlah lagu dalam playlist: "))

#Loop untuk memasukan lagu ke dalam circular linked list
for i in range(1, num_songs + 1):
  titel = input(f"Masukkan judul lagu ke-{i}: ")
  playlist.add_song(titel)

#Menampilkan daftar lagu
playlist.show_playlist()

#Menu navigasi pengguna
while True:
  command = input("\n Tekan 'n' untuk memutar lagu berikutnya, tekan 'x' untuk keluar.\n Masukan perintah: ").lower()
  if command == 'n':
    playlist.play_next()
  elif command == 'x':
    print ("\n Keluar dari pemutar musik")
    break
  else:
    print("\n Perintah tidak dikenal")
````

````python
#Operasi dasar pada linked list
#Definisi Node dalm Singly Linked List
class Node:
  def __init__(self, data):
    self.data = data #Menyimpan nilai node
    self.next = None #Pointer ke node berikutnya

#Definisi Singly Linked List dengan berbagai operasi
class LinkedList:
  def __init__(self):
    self.head = None #Awalnya, daftar kosong

  #Menambahkan elemen diawal
  def add_at_beginning(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

  #Menambahkan elemen di akhir
  def add_at_end(self, data):
      new_node = Node(data)
      if not self.head:
        self.head = new_node
        return
      temp = self.head
      while temp.next:
        temp = temp.next
      temp.next = new_node

  #Menambahkan Elemen di posisi tertentu
  def add_at_position(self, position, data):
    if position < 0:
        print("Posisi tidak valid")
        return
    new_node = Node(data)
    if position == 0:
        new_node.next = self.head
        self.head = new_node
        return
    temp = self.head
    for _ in range(position -1):
      if not temp:
        print("Posisi melebihi panjang daftar!")
        return
      temp = temp.next
    new_node.next = temp.next
    temp.next = new_node

  #Menghapus elemen diawal
  def delete_at_beginning(self):
      if not self.head:
          print("Daftar kosong, tidak ada yang bisa dihapus.")
          return
      self.head = self.head.next

  #Menghapus elemen diakhir
  def delete_at_end(self):
      if not self.head:
          print("Daftar kosong, tidak ada yang bisa dihapus.")
          return
      if not self.head.next:
          self.head = None
          return
      temp = self.head
      while temp.next.next:
        temp = temp.next
      temp.next = None

  #Menghapus elemen di posisi tertentu
  def delete_at_position(self, position):
      if not self.head:
          print("Daftar kosong, tidak ada yang bisa dihapus.")
          return
      if position == 0:
          self.head = self.head.next
          return
      temp = self.head
      for _ in range(position - 1):
          if not temp or not temp.next:
              print("⚠️Posisi tidak ditemukan!")
              return
          temp = temp.next
      temp.next = temp.next.next

  #Mencari elemen dalam linked list
  def search(self, key):
      temp = self.head
      while temp:
        if temp.data == key:
            print(f"Elemen {key} ditemukan dalam daftar.")
            return
        temp = temp.next
      print(f"Elemen {key} tidak ditemukan dalam daftar.")

  #Membalik (reverse) Linked List
  def reverse(self):
      prev = None
      current = self.head
      while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
      self.head = prev

  #Menampilkan Linked List
  def display(self):
      if not self.head:
          print("Daftar kosong!")
          return
      temp = self.head
      while temp:
          print(temp.data, end=" -> ")
          temp = temp.next
      print("None")

# ========== Program Utama (main) ========== #

#Membuat objek linked list
# Moved this part outside the class definition
il = LinkedList()

#Meminta pengguna memasukkan elemen awal
num_elements = int(input("Masukkan jumlah awal elemen dalam daftar: "))
for i in range (num_elements):
    data = int(input(f"Masukkan elemen ke-{i+1}: "))
    il.add_at_end(data)

#Menu Operasi linked list
while True:
    print("\nPilih Operasi:")
    print("1. Tambahkan elemen diawal")
    print("2. Tambahkan elemen di akhir")
    print("3. Tambahkan elemen di posisi tertentu")
    print("4. Hapus elemen diawal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balik (reverse) daftar")
    print("9. Tampilkan daftar")
    print("10. Keluar")

    choise = int(input("Masukkan pilihan : "))

    if choise == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        il.add_at_beginning(data)
    elif choise == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        il.add_at_end(data)
    elif choise == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        il.add_at_position(position, data)
    elif choise == 4:
        il.delete_at_beginning()
    elif choise == 5:
        il.delete_at_end()
    elif choise == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        il.delete_at_position(position)
    elif choise == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        il.search(key)
    elif choise == 8:
        il.reverse()
        print("Daftar berhasil dibalik!")
    elif choise == 9:
        il.display()
    elif choise == 10:
        print("Terima kasih! Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid. Silakan coba lagi ")
````

Dalam laporan, tuliskan penjelasan mengenai implementasi oprasi dasar pada Linked List dalam kode program diatas diterapkan

Jawab : Singly Linked List merupakan struktur data yang terdiri dari serangkaian node yang menyimpan data dan memiliki referensi ke node berikutnya. Penjelasan implementasi operasi dasar yang telah diterapkan dalam kode:

1. Menambahkan Elemen
  a. Menambahkan elemen di awal (add_at_beginning)
    digunakan untuk menambahkan elemen baru di awal Linked List
    * Membuat node baru (new_node) dengan data yang diberikan
    * Menjadikan next dari (new_node) sebagai head saat ini.
    * Memperbarui head menjadi new_node.
    *  Kompleksitas Waktu: O(1) – karena hanya membutuhkan perubahan referensi.
    ```
    def add_at_beginning(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node
    ```

  b. Menambahkan Elemen di Akhir (add_at_end)
    digunakan untuk menambahkan elemen baru di akhir linked list
    * Membuat node baru (new_node) dengan data yang diberikan.
    * Jika head kosong, node baru menjadi head.
    * Jika tidak, iterasi sampai node terakhir (temp.next == None), lalu hubungkan next dari node terakhir ke new_node.
    * Kompleksitas Waktu: O(n) – karena harus melakukan traversal hingga node terakhir.
    ```
    def add_at_end(self, data):
    new_node = Node(data)
    if not self.head:
        self.head = new_node
        return
    temp = self.head
    while temp.next:
        temp = temp.next
    temp.next = new_node
    ```

  c.  Menambahkan Elemen di Posisi Tertentu (add_at_position)
    memungkinkan penambahan elemen di posisi yang ditentukan
    * Jika posisi 0, panggil add_at_beginning.* Lakukan traversal hingga posisi yang ditentukan.
    * Selipkan new_node di posisi tersebut.
    *  Kompleksitas Waktu: O(n) – karena harus melakukan traversal hingga posisi tertentu.
    ```
    def add_at_position(self, position, data):
    if position < 0:
        print("Posisi tidak valid")
        return
    new_node = Node(data)
    if position == 0:
        new_node.next = self.head
        self.head = new_node
        return
    temp = self.head
    for _ in range(position - 1):
        if not temp:
            print("Posisi melebihi panjang daftar!")
            return
        temp = temp.next
    new_node.next = temp.next
    temp.next = new_node

    ```

2.  Menghapus Elemen
  a. Menghapus Elemen di Awal (delete_at_beginning)
    Menghapus elemen pertama dengan:
    * Mengubah head menjadi head.next.
    * Kompleksitas Waktu: O(1) – karena hanya mengubah referensi head.
    ```
    def delete_at_beginning(self):
    if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus.")
        return
    self.head = self.head.next
    ```

  b. Menghapus Elemen di Akhir (delete_at_end)
    Menghapus elemen terakhir dengan:
    * Jika hanya ada satu elemen, head diubah menjadi None.
    * Jika lebih dari satu elemen, traversal sampai sebelum elemen terakhir (temp.next.next == None), lalu temp.next = None
    * Kompleksitas Waktu: O(n) – karena harus mencari elemen terakhir.
    ```
    def delete_at_end(self):
    if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus.")
        return
    if not self.head.next:
        self.head = None
        return
    temp = self.head
    while temp.next.next:
        temp = temp.next
    temp.next = None

    ```

  c. Menghapus Elemen di Posisi Tertentu (delete_at_position)
    Menghapus elemen pada posisi tertentu dengan:
    * Jika posisi 0, panggil delete_at_beginning.* Lakukan traversal sampai posisi sebelum elemen yang akan dihapus, lalu hubungkan next dari node sebelumnya ke next.next.* Kompleksitas Waktu: O(n) – karena harus melakukan traversal.
    ```
    def delete_at_position(self, position):
    if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus.")
        return
    if position == 0:
        self.head = self.head.next
        return
    temp = self.head
    for _ in range(position - 1):
        if not temp or not temp.next:
            print("⚠️ Posisi tidak ditemukan!")
            return
        temp = temp.next
    temp.next = temp.next.next

    ```

3. Mencari Elemen (search)
  a. Metode ini mencari nilai dalam linked list:
  * Traversal satu per satu hingga menemukan nilai yang dicari.
  * Kompleksitas Waktu: O(n) – karena bisa saja harus mencari sampai elemen terakhir.

  def search(self, key):

  a. Membalik urutan linked list dengan:
  * Menggunakan tiga pointer (prev, current, next_node).
  * Mengubah arah setiap next agar menunjuk ke elemen sebelumnya.
  * Kompleksitas Waktu: O(n) – karena harus mengunjungi semua node.

Kesimpulan
1. Linked List memungkinkan manipulasi data secara dinamis tanpa perlu mendeklarasikan ukuran awal seperti array.
2. Operasi penambahan dan penghapusan cepat di awal, tetapi lambat di tengah dan akhir karena memerlukan traversal.
3. Metode reverse mengubah urutan elemen, berguna dalam banyak skenario.
4. Setiap operasi memiliki kompleksitas waktu yang berbeda, yang harus dipertimbangkan saat memilih struktur data yang sesuai.
